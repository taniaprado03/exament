```python
opc = 0
lista_asientos = []
lista_pasajeros = []
nombre_pasajero = ''
rut_pasajero = ''
total=0

def ValidarNumero():
    opc = input('Ingresa una opcion: ')
    while not opc.isnumeric():
        print('La opcion ingresada tiene que ser numérico')
        opc = input('Ingresa un valor: ')
    return int(opc)

def CargarAsientos():
    for x in range(1, 101, 1):
        lista_asientos.append(x)
CargarAsientos()

def VerAsientos():
    contador_1 = 0
    contador_2 = 0

    print('''
      • Platinum, $120.000. (1 - 20)
      • Gold $80.000. (21 - 50)
      • Silver $50.000. (51-100)

    ''')
    for x in lista_asientos:
        contador_1 = contador_1 + 1
        contador_2 = contador_2 + 1

        if contador_2 == 51:
            print('-' * 40)
            print('-' * 40)

        if contador_1 < 10:
            print('|', end='')
            print(x, end='|')
        else:
            print(x, end='|')
            print('')
            contador_1 = 0

def ComprarAsiento():
    mostrar_asientos = VerAsientos()
    print('N° de Asiento:')
    numero_asiento = ValidarNumero()




    print('¿Está seguro de realizar esta compra? (1.- Si / 2.- No)')
    opc = ValidarNumero()

    if opc == 1:
        nombre_pasajero = input('Nombre: ')
        rut_pasajero = input('Rut (ejem: 111111119): ')
        while len(rut_pasajero) < 8 or len(rut_pasajero) > 9:
            print('El valor del rut no puede ser menor a 8 caracteres o superior a 9')
            rut_pasajero = input('Rut (ejem: 111111119): ')

    lista_pasajeros.append([numero_asiento, nombre_pasajero, rut_pasajero])
    lista_asientos[(numero_asiento - 1)] = 'X'

def Mostrar_listado_asistentes():
    print("Listado de asistentes:")
    for lista_pasajeros in lista_pasajeros:
        print(f"RUN: {lista_pasajeros['run']}, Asiento: {lista_pasajeros['ubicacion']}")



while True:
    print('''
    1. Comprar entrada
    2. Ver asientos disponibles
    3. Ver listado de asistente
    4. Mostrar ganancias totales
    5. Salir
    ''')

    opc = ValidarNumero()

    if opc == 1:
        ComprarAsiento()

    elif opc == 2:
        VerAsientos()
    elif opc == 3:
        Mostrar_listado_asistentes()



    elif opc == 5:
        print('Salir')
        break



    print(lista_pasajeros)


```
